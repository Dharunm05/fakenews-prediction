```python
from tensorflow.keras import layers
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size=5000
```

```python
msg=x.copy()
```

```python
msg["title"]
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
p=PorterStemmer()
corpus=[]
for title in msg["title"]:
  review=re.sub("[^a-zA-Z]","  ",title)
  review=review.lower()
  review=review.split()
  review=[p.stem(word) for word in review if not word in stopwords.words("english")]
  review=" ".join(review)
  corpus.append(review)
```

```python
corpus
```

```python
onerepr=[one_hot(words,voc_size) for words in corpus]
```

```python
onerepr
```

```python
sent_len=20
embedded_docs=pad_sequences(onerepr,padding="pre",maxlen=sent_len)
embedded_docs
```

```python
embedded_docs[0]
```

```python
model=Sequential()
embedding_vector_features=40
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_len))
model.add(LSTM(100))
model.add(Dense(1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
print(model.summary())
```

```python
import numpy as np
x_f=np.array(embedded_docs)
y_f=np.array(y)
```

```python
x_f.shape
```

```python
y_f.shape
```

```python
from tensorflow.keras.layers import Dropout
model=Sequential()
embedding_vector_features=40
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_len))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dense(1,activation="sigmoid"))
model.add(Dropout(0.3))
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
print(model.summary())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_f,y_f,test_size=0.33,random_state=42)
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(y_test,y_pred.round())
```

```python
accuracy_score(y_test,y_pred.round())
```

```python
classification_report(y_test,y_pred.round())
```

```python
# prompt: i want add input from the user

def predict_user_input(user_input):
  review = re.sub("[^a-zA-Z]", "  ", user_input)
  review = review.lower()
  review = review.split()
  review = [p.stem(word) for word in review if not word in stopwords.words("english")]
  review = " ".join(review)
  one_hot_rep = one_hot(review, voc_size)
  embedded_doc = pad_sequences([one_hot_rep], padding="pre", maxlen=sent_len)
  prediction = model.predict(embedded_doc)
  return prediction

user_input = input("Enter a news title: ")
prediction = predict_user_input(user_input)

if prediction[0][0] >= 0.5:
  print("The news is likely real.")
else:
  print("The news is likely fake.")
```
